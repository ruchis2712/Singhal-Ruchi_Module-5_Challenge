```python
%matplotlib inline
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import statistics
import math
import datetime
```

```python
# Read Data File

pyber_city_data=pd.read_csv("Resources/city_data.csv")
pyber_ride_data=pd.read_csv("Resources/ride_data2.csv")

pyber_ride_data.isnull().sum()
```

```python
pyber_city_data['type'].value_counts()
```

```python
pyber_data_df=pd.merge(pyber_ride_data,pyber_city_data, how="left", on=['city','city'])
```

```python
# Create DF for each city type

urban_cities_df = pyber_data_df[pyber_data_df["type"] == "Urban"]
suburban_cities_df = pyber_data_df[pyber_data_df["type"] == "Suburban"]
rural_cities_df = pyber_data_df[pyber_data_df["type"] == "Rural"]
```

```python
# Get the number of rides for cities
urban_ride_count = urban_cities_df.groupby(["city"]).count()["ride_id"]
suburban_ride_count=suburban_cities_df.groupby(["city"]).count()["ride_id"]
rural_ride_count=rural_cities_df.groupby(["city"]).count()["ride_id"]
```

```python
pyber_data_df.dtypes
```

```python
# Get the average fare for each city

urban_avg_fare = urban_cities_df.groupby(["city"]).mean()["fare"]
suburban_avg_fare=suburban_cities_df.groupby(["city"]).mean()["fare"]
rural_avg_fare=rural_cities_df.groupby(["city"]).mean()["fare"]
```

```python
# Get the driver count for each city

urban_driver_count = urban_cities_df.groupby(["city"]).mean()["driver_count"]
suburban_driver_count=suburban_cities_df.groupby(["city"]).mean()["driver_count"]
rural_driver_count=rural_cities_df.groupby(["city"]).mean()["driver_count"]
```

```python
# Create Bubble Chart for All cities

# Create Bubble Chart for URBEN cities
plt.scatter(urban_ride_count,urban_avg_fare, s=10*urban_driver_count, color="coral", label="Urban", edgecolor="black", linewidth=1, alpha=.8)


# Create Bubble Chart for Suburban cities
plt.scatter(suburban_ride_count,suburban_avg_fare, s=10*suburban_driver_count, color="skyblue", label="Suburban", edgecolor="black", linewidth=1, alpha=.8)

# Create Bubble Chart for Rural cities
plt.scatter(rural_ride_count,rural_avg_fare, s=10*rural_driver_count, color="gold", label="Rural", edgecolor="black", linewidth=1, alpha=.8)

# Add other chart properties

plt.xlabel("Total Number of Rides (per City)", fontsize=12)
plt.ylabel("Avg. Fare ($)", fontsize=12)
plt.title("PyBer Ride-Sharing Data (2019)", fontsize=20)
plt.legend()
plt.grid()

# Formatting the legend

# Create a legend
lgnd = plt.legend(fontsize="12", mode="Expanded",
         scatterpoints=1, loc="best", title="City Types")
lgnd.legendHandles[0]._sizes = [75]
lgnd.legendHandles[1]._sizes = [75]
lgnd.legendHandles[2]._sizes = [75]
lgnd.get_title().set_fontsize(12)

# Add text box message

# Incorporate a text label about circle size.
plt.text(42, 35, "Note:\nCircle size correlates\nwith driver count per city.", fontsize="12")

# Save chart to file

plt.savefig("Analysis/Fig1.png")

plt.show()
```

```python
# Import mpl to change the plot configurations using rcParams.
import matplotlib as mpl

# PIE CHART for % Fares by City Type

total_fares=pyber_data_df.groupby(["type"]).sum()["fare"]
x_labels=total_fares.keys()

explode = [0, 0, 0.1]
colors = ["gold", "lightskyblue", "lightcoral"]
plt.subplots(figsize=(8,8))

plt.pie(total_fares, labels=x_labels, autopct="%1.1f%%", explode=explode, colors=colors, startangle=150,shadow=True)

plt.title("% of Total Fares by City Type")

# Change the default font size from 10 to 14.
mpl.rcParams['font.size'] = 14

# Save and Show Figure
plt.savefig("analysis/Fig5.png")

plt.show()
```

```python
# PIE CHART for total rides by city type

ride_count=pyber_data_df.groupby(["type"]).count()["ride_id"]
x_labels=ride_count.keys()

explode = [0, 0, .1]
colors = ["gold", "lightskyblue", "lightcoral"]
plt.subplots(figsize=(8,8))

plt.pie(ride_count, labels=x_labels, autopct="%1.1f%%", explode=explode, colors=colors, startangle=150,shadow=True)

plt.title("% of Total Rides by City Type")

# Change the default font size from 10 to 14.
mpl.rcParams['font.size'] = 14

# Save and Show Figure
plt.savefig("analysis/Fig6.png")

plt.show()
```

```python
# PIE CHART for % Drivers by city type

driver_count=pyber_data_df.groupby(["type"]).sum()["driver_count"]
x_labels=ride_count.keys()

explode = [0, 0, .1]
colors = ["gold", "lightskyblue", "lightcoral"]
plt.subplots(figsize=(8,8))

plt.pie(driver_count, labels=x_labels, autopct="%1.1f%%", explode=explode, colors=colors, startangle=150,shadow=True)

plt.title("% of Total Rides by City Type")

# Change the default font size from 10 to 14.
mpl.rcParams['font.size'] = 14

# Save and Show Figure
plt.savefig("analysis/Fig7.png")

plt.show()
```

```python
avg_fare_per_ride=total_fares/ride_count
avg_fare_per_driver=total_fares/driver_count

final_df= pd.DataFrame({
    "Total Rides":ride_count,
    "Total Drivers":driver_count,
    "Total Fare":total_fares,
    "Avg. Fare per Ride":avg_fare_per_ride,
    "Avg. Fare per Driver":avg_fare_per_driver

})
final_df["Total Fare"]=final_df["Total Fare"].map("${:,.2f}".format)
final_df["Avg. Fare per Ride"]=final_df["Avg. Fare per Ride"].map("${:.2f}".format)
final_df["Avg. Fare per Driver"]=final_df["Avg. Fare per Driver"].map("${:.2f}".format)
final_df.to_csv("analysis/Pyber_Summary_DataFrame.csv")
final_df
```

```python
# Rename column and reset index of the main merged dataframe

pyber_data_df=pyber_data_df.rename(columns={'city': 'City', 'date':'Date','fare':'Fare', 'ride_id': 'Ride Id',
                                            'driver_count': 'No. Drivers', 'type':'City Type'})
```

```python
# Make a copy of the merged dataframe
pyber_df_copy=pyber_data_df[['Date','City Type','Fare']].copy()

# Set Date as Index
pyber_df_copy.set_index('Date', inplace=True)

# Convert Index into DateTime
pyber_df_copy.index=pd.to_datetime(pyber_df_copy.index)

# Check Index type
pyber_df_copy.info()
```

```python
# Create groups for each city type
city_type_group= pyber_df_copy.groupby(['City Type'])
urban_group=city_type_group.get_group('Urban')
suburban_group=city_type_group.get_group('Suburban')
rural_group=city_type_group.get_group('Rural')
```

```python
# Calculate Sum of Fares and create series for each city type

urban_fare_by_month=urban_group.groupby(urban_group.index).sum()['Fare']
suburban_fare_by_month=suburban_group.groupby(suburban_group.index).sum()['Fare']
rural_fare_by_month=rural_group.groupby(rural_group.index).sum()['Fare']
```

```python
pivot_df=pd.DataFrame({
    "Urban Fares":urban_fare_by_month,
    "Suburban Fares":suburban_fare_by_month,
    "Rural Fares":rural_fare_by_month
})

pivot_df.sort_index(inplace=True)

# retrieving rows by loc method 
selected_rows = pivot_df.loc['2019-01-01':'2019-04-28']
```

```python
urban=selected_rows.resample('M').sum()['Urban Fares']
suburban=selected_rows.resample('M').sum()['Suburban Fares']
rural=selected_rows.resample('M').sum()['Rural Fares']
urban
```

```python
# use Matplotlib style (stylesheets) fivethirtyeight
plt.style.use('fivethirtyeight')

# Plot the line charts
x_axis=urban.index.month_name()

fig, ax = plt.subplots(figsize=(15,8))
ax.plot(x_axis,urban, color='gold',  linewidth=2, label='Urban')
ax.plot(x_axis,suburban, color='orange',  linewidth=2, label='Suburban')
ax.plot(x_axis,rural,  color='teal',  linewidth=2, label='Rural')

# chart title
ax.set_title("Monthly Fare by City Type",fontsize=20)
# Create labels for the x and y axes.
ax.set_xlabel("Month",fontsize=14)
ax.set_ylabel("Fare($ USD)",fontsize=14)
# Add the legend
ax.legend()

# CHANGE MAJOR TICKS
ax.set_yticks(np.arange(0, 10500, step=500.0))

# Formatting the legend

# Create a legend
lgnd = plt.legend(fontsize="12", mode="Expanded",
         scatterpoints=1, loc="center", title="City Type")
lgnd.legendHandles[0]._sizes = [75]
lgnd.legendHandles[1]._sizes = [75]
lgnd.legendHandles[2]._sizes = [75]
lgnd.get_title().set_fontsize(12)


# Save to file
plt.savefig("analysis/monthy_fares_by_CityType.png")



plt.show()
```
